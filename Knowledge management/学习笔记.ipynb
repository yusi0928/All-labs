## Use if statement  用if表达，让负数变正数

```python
import math

def sqrt_for_all(x):

    if x < 0:
        x = x*-1
    return math.sqrt(x)

sqrt_for_all(-2)
```

## 如果除数是0，返回值为0

```python
import math

def divide(x, y):

    if y == 0:
        return 0
    else:
        return x / y

divide(5, 0)
```

## 自定义Errors的内容

```python
def log_square(x):

    if x == 0:
        raise ValueError("The number entered cannot be zero")
    else:
        return math.log(x**2)
    
print(log_square(5))
print(log_square(0))




import re

def check_capital(x):
    if re.search('[A-Z]',x) is None:    # re.search 返回值是match的词，没有match的词就是none
        raise ValueError("No capital letters detected")
    else:
        return True

print(check_capital("John"))
print(check_capital('john'))
```

## re.sub('目标词','替代词', text)拆长句子，然后全部小写，并数出现的数量。 可以应用于政策分析，解读十五大报告的关键词

```python
import re

poem = """Some say the world will end in fire,
Some say in ice.
From what I’ve tasted of desire
I hold with those who favor fire.
But if it had to perish twice,
I think I know enough of hate
To say that for destruction ice
Is also great
And would suffice."""


dictionary = {}
poem_w = poem.split()   # 这是个list，把list的每个词隔开， 有的词是和line breaks在一起的
print(poem_w)

for i in poem_w:
    clean_word = re.sub('\W','',i).lower()   # \W=[^a-zA-Z0-9_] 所有不是alphabeic or numerical 形式的字符=就是那些特殊字符          
    dictionary[clean_word] = poem_w.count(i)    # 添加到dic： dic[key]='value'
    
print(dictionary)
```

## 如果可以相除的话，返回值1，如果不能相除，返回值0

```python
mod = lambda num, denom: 1 if (num % denom == 0) else 0  #num % denom == 0

def divisor(b):
    return lambda a: mod (a, b)


divisible5 = divisor(5)

divisible5(10)
```

## lambda和zip，sorted

```python
list1 = ['Green', 'cheese', 'English', 'tomato']
list2 = ['eggs', 'butter', 'cucumber', 'tomato']

Total = [x for x in zip(list1,list2)]  # 第一个x是('cheese', 'butter')
print(Total)   # 返回是一个由tuple组成的list


sortedList = sorted(Total,key=lambda x: x[1][0])  #以每个tuple里第2个单词的第1个字母顺序排序

print(sortedList)
```

## Earthquark sample

```python

import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.emsc-csem.org/Earthquake/'
html = requests.get(url).content
soup = BeautifulSoup(html, "html.parser")


#tbody is the whole body of earthquakes and each tr refers to every earthquake
earthquakes = soup.find('tbody', {'id': 'tbody'}).find_all("tr")

number_of_results = 20   # 只提取前20个地震信息
latest_earthquakes = []


print(earthquakes[0])    #Let's analyze the first object


for earthquake in earthquakes[:number_of_results]:  
    
    date,time = earthquake.find('td', {'class': 'tabev6'}).find('a').text.split()  # Double variables
    
# Latitude and longitude
# First I get the lat and longitude degrees (they are in two different but with the same name tags)
    lat_deg, lon_deg = earthquake.find_all('td', {'class': 'tabev1'})             # 提取经纬度
    lat_dir, lon_dir, magnitude = earthquake.find_all('td', {'class': 'tabev2'})  # 提取东南西北方向
    lat_deg = f"{lat_deg.text.strip()} {lat_dir.text.strip()}"                    # 合并成准确位置
    lon_deg = f"{lon_deg.text.strip()} {lon_dir.text.strip()}"
    
    region = earthquake.find('td', {'class': 'tb_region'}).text.strip()

    earthquake_summary = [date, time, lat_deg , lon_deg, region]   # 里面每个element都由string组成，不存在list of list的问题
    latest_earthquakes.append(earthquake_summary)                  # 把每一个地震信息收到list里，收20条地震
   
df = pd.DataFrame(latest_earthquakes, columns=['Date', 'Time', 'Latitude', 'Longitude', 'Region'])
df
```

A confidence interval gives us information about the true mean of the population. So for a 95% confidence interval, we are 95% sure that the mean of the population is within the confidence interval.

### Case 1: 
We want to estimate the average size of the men of a country with a confidence level of 80%. Assuming that the standard deviation of the sizes in the population is 4, get the confidence interval with a sample of men selected randomly, whose heights are:

heights = [167, 167, 168, 168, 168, 169, 171, 172, 173, 175, 175, 175, 177, 182, 195]

Step 1: confidence interval with a sample of men

### Calculate the standard deviation of the sample, you will need the standard deviation of the whole poplulation

```python
from scipy.stats import norm
import numpy as np
import statistics
import math

sample=[167, 167, 168, 168, 168, 169, 171, 172, 173, 175, 175, 175, 177, 182, 195]

population_std=4   # the standard deviation of the population
sample_mean=np.mean(sample)
sample_std=population_std/math.sqrt(len(sample))  

norm.interval(0.8,loc=sample_mean,scale=sample_std)
```

## chi2.interval(conficenceLevel, df)

Step 2: calculate a confidence interval for the variance at 90% level.

```python
from scipy.stats import chi2

sample = [167, 167, 168, 168, 168, 169, 171, 172, 173, 175, 175, 175, 177, 182, 195]

df = len(sample) - 1
score = chi2.interval(0.90, df)
sample_quasi_var = np.var(sample)

conf_int = [ df * sample_quasi_var / score[1], df * sample_quasi_var / score[0]]
print(f"Confidence interval 90%: {conf_int}")
```

# ---------------Hypothesis Tests------------------------

### one-sided  VS  two-sided P-value

It is possible to reject a value in a two tailed test that you would accept 
in a one tailed test. Why? 

Because 
#### the two tailed test is just testing if the mean is different: 
it could be different by being EITHER greater or smaller. 

In the one tailed test, we are testing already if it is greater or smaller. 
For that reason, the p-value is half that of the two sided test.

## stat,pval=ttest_1samp(series,NullHypothesisValue)

### Case 1: 
Test whether the hourly wage of all hourly workers is significantly different from $30/hr. Import the correct one sample test function from scipy and perform the hypothesis test for a 95% two sided confidence interval.

```python
from scipy.stats import ttest_1samp

salaries=pd.read_csv('/Users/abc/Portfolio/module-2/2.7 Hypothesis-Testing/employees chicago.csv')


# Ho: the mean hourly wage == $30/h.
# Ha: the mean hourly wage != $30/h.


hourly_rate = salaries[salaries['Salary or Hourly'] == 'Hourly']['Hourly Rate']

stat,pval=ttest_1samp(hourly_rate,30)   # Default confidence level 95%


print("Null hypothesis rejected") if p < 0.05 else print("Null hypothesis can't be rejected")

print('2 sided P-value is: {}'.format(pval))
print('stat is: {}'.format(stat))
```

### Case 2: 
The chief of police in Chicago claimed in a press briefing that salaries this year are higher than last year's mean of $86000/year a year for all salaried employees. Test this one sided hypothesis using a 95% confidence interval.

```python
# Ho: the salaries this year <= last year's mean of $86000/year for salaried employees.
# Ha: the salaries this year > last year's mean of $86000/year for salaried employees.


police_filtered = salaries[(salaries['Salary or Hourly'] == 'Salary') & (salaries['Department'] == 'POLICE')]
police_salaries = police_filtered['Annual Salary']


stat, p_two_tailed = ttest_1samp(police_salaries, 86000)  # Default confidence level 95%


p_one_tailed = p_two_tailed / 2

if p_one_tailed < 0.05 and stat > 0:     # stat >0 ????
    print("Null hypothesis rejected")
    print("The mean salary this year is greater than last year's mean")
else:
    print("Null hypothesis can't be rejected")
    
print('Statistic: ', stat)
print('p-value: ', p_one_tailed)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from pandas.plotting import lag_plot
from statsmodels.tsa.stattools import adfuller
import warnings
warnings.filterwarnings("ignore")

sensor = pd.read_csv('/Users/abc/All Labs/module-2/4.2 lab-time-series/sub_1.csv', index_col=0)

############################################# Data preparation #############################################
# change the index to a time series index. Create a time series index starting at 1/1/2018 00:00:00 and ending at 1/1/2018 00:29:10. The number of periods is equal to the number of rows in sensor
index = pd.date_range('1/1/2018 00:00:00', '1/1/2018 00:29:10', periods=sensor.shape[0])
sensor.index = index   # Assign the time series index to the dataframe's index.


# decompose the time series and evaluate the patterns in the data. Load the statsmodels.api submodule and plot the decomposed plot of userAcceleration.x. Set freq=60 in the seasonal_decompose function.
results = sm.tsa.seasonal_decompose(sensor['userAcceleration.x'], freq=60)
results.plot()
results = sm.tsa.seasonal_decompose(sensor['rotationRate.x'], freq=60)
results.plot()





################################################ Model assumption ################################################
lag_plot(sensor['userAcceleration.x'])
plt.show()
lag_plot(sensor['rotationRate.x'])
plt.show()

'''
Autocorrelation in time series means that the series is linearly related to a lagged version of itself. 
For features 'rotationRate.x' and 'userAcceleration.x', the observations at time t have a positive linear 
correlation with the observations at time t+1. One of the assumptions of regression analysis is that the
data has no autocorrelation. If we try to do a regression analysis on data with autocorrelation, 
then our analysis will be misleading.
'''

# test both variables for stationarity. Perform the Augmented Dickey Fuller test
# Ho: userAcceleration.x is not stationary
# Ha: userAcceleration.x is stationary

if adfuller(sensor['userAcceleration.x'])[1] < 0.05:
    print("We can reject the null hypothesis with a 95% confidence level.")
else:
    print("We can't reject the null hypothesis.")


# Ho: rotationRate.x is not stationary
# Ha: rotationRate.x is stationary

if adfuller(sensor['rotationRate.x'])[1] < 0.05:
    print("We can reject the null hypothesis with a 95% confidence level.")
else:
    print("We can't reject the null hypothesis.")
    
    
'''
ARMA models work on the assumption that the time series is stationary. Stationarity means that the statistical 
properties of a process generating a time series do not change over time. It does not mean that the series does 
not change over time, just that the way it changes does not itself change over time. In both cases, we can reject 
the null hypothesis that the time series is not stationary, so we can create an ARMA model.

'''

############################################ Model building ###############################################
from statsmodels.tsa.arima_model import ARMA
# ARMA of order (p,q)
# p: order of the autoregressive model (AR).
# q: order of the moving average model (MA).


# Train and test set
index = sensor.shape[0] - 10
train = sensor['userAcceleration.x'][:index] 
test = sensor['userAcceleration.x'][index:]
print(f"The number of samples in train is {train.shape[0]} and the number of samples in test is {test.shape[0]}.")

# select Model ARMA
model = ARMA(train, order=(2,1), freq='S')  
arma = model.fit(disp=False)

# Predictions
predictions = arma.predict(test.index[0], test.index[-1])

# evaluation
from statsmodels.tools.eval_measures import rmse
rmse(test, predictions)          # RMSE= Root Mean Squared Error

'''
The RMSE is the square root of the variance of the residuals. It indicates the absolute fit of the model 
to the data, so how close the observed data points are to the model’s predicted values. Whereas R-squared 
is a relative measure of fit, RMSE is an absolute measure of fit. As the square root of a variance, RMSE 
can be interpreted as the standard deviation of the unexplained variance, and has the useful property of 
being in the same units as the response variable. Lower values of RMSE indicate better fit. RMSE is a good 
measure of how accurately the model predicts the response, and it is the most important criterion for fit 
if the main purpose of the model is prediction. We can say that the deviation from the true prediction is 
+/- 0.22
'''
```

## Data cleaning

```python


import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer
import random


def clean_up(s):      
    s_1 = re.sub('\d', '', s)                         # Step 1 - Remove all digits
    s_2 = re.sub('(http|https)://([^\s]+)', '', s_1)  # Step 2 - Remove URL
    s_3 = re.sub('\W', ' ', s_2 )                     # Step 3 - Remove all special chars
    s_4 = (s_3.lower())                               # Step 4 - to lower and split to list
    return s_4


def tokenize(s):    
    return nltk.word_tokenize(s)



def stem_and_lemmatize(l):
    lemmatizer = WordNetLemmatizer()
    ps = PorterStemmer()
    result = []
    for w in l:
        result.append(lemmatizer.lemmatize(ps.stem(w)))
    return result


def remove_stopwords(l):
    stop_words = stopwords.words('english') 
    return [word for word in l if word not in stop_words]
```

## Data importing and sampling

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")

cols = ['target','id','data','flag','user','text']
data=pd.read_csv('/Users/abc/Desktop/vehicles/training.1600000.processed.noemoticon.csv', engine='python',header=None, names=cols)

data.shape
# data_sample = data[80000:810000]   # 取一定区间的值做sample，这样就不用打开全部数据了，数据太大电脑无能
sample=data.sample(n=10000, axis=0)  # 或者任意sampling

data_sample.shape

data['target'].value_counts()
data_sample['target'].value_counts()
data_sample.head()


# Applying functions to clean texts
data_sample['text_processed'] = data_sample.apply(lambda x: clean_up(x['text']), axis=1)
data_sample['text_processed'] = data_sample.apply(lambda x: tokenize(x['text_processed']), axis=1)
data_sample['text_processed'] = data_sample.apply(lambda x: stem_and_lemmatize(x['text_processed']), axis=1)
data_sample['text_processed'] = data_sample.apply(lambda x: remove_stopwords(x['text_processed']), axis=1)

data_sample.head()
```

## Creating Bag of Words

### The purpose of this step is to create a bag of words from the processed data. The bag of words contains all the unique words in your whole text body (a.k.a. corpus) with the number of occurrence of each word. It will allow you to understand which words are the most important features across the whole corpus.

A convenient library to calculate the term frequency distribution is NLTK's FreqDist class (documentation). Then select the top 100 words from the frequency distribution.

```python
from nltk.probability import FreqDist


words = [w for lst in data_sample['text_processed'] for w in lst]  # Iterating over each list on text processed to extract all the words

fdist = FreqDist(words)  # Passing the previous variable to FreqDist function

# Using the .most_common function to extract the top 100 words
# After running it, modify this value to 100 to be able to run it
top100 = fdist.most_common(100)

# The previous variable returns a tuple containing the word and the freq, we are now extracting all the words
top_words = [w[0] for w in top100]
```

## Building Features

```python
#Creating final list
final_data = []


#iterrows returns index + row 
#Sliced the dataframe otherwise is heavy to run
for i, r in data_sample[8000:12000].iterrows():
    
    #One dict per row
    result = {}
    
    # Iterating over the top5000 words and then checking if the word in word500 is in the column 'text_processed'
    for w in top_words:
        result[w] = (w in r['text_processed'])
        
    # If the value on the column target for each row is 0 means it is negative, otherwise (4) is positive
    if r['target'] == 0:
        # Negative
        sentiment = False 
    else:
        # positive
        sentiment = True
        
    #Saving per each row the dict and the sentiment
    final_data.append((result, sentiment))

print(len(final_data))
final_data[0]
```

## Testing Naive Bayes Model

```python
# Shuffle list
random.shuffle(final_data)

# Split 80/20
training_data = final_data[:3200]
test_data = final_data[3200:]


classifier = nltk.NaiveBayesClassifier.train(training_data)

classifier.show_most_informative_features(10)
```

```python
accuracy=nltk.classify.accuracy(classifier, test_data)
print("Accuracy score:",accuracy)
```
