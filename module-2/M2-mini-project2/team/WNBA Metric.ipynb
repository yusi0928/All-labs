```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("wnba_clean.csv")
```

```python
data.columns
```

```python

pd.set_option('display.max_columns', None)
data["PpG"]=data["PTS"]/data["Games Played"]
data["ApG"]=data["AST"]/data["Games Played"]
data["RpG"]=data["REB"]/data["Games Played"]
data["SpG"]=data["STL"]/data["Games Played"]
data["BpG"]=data["BLK"]/data["Games Played"]
data["TpG"]=data["TO"]/data["Games Played"]
data.describe()
```

```python
"""""
IER (Ironhack Efficency Rating)
should include:
1. Points per game (mean = 7, max=20)      counts: 20%
2. Ass per Game (mean=1,6, max=8)        counts: 5%
3. Reb per Game (mean=3, max=11)        counts:10%
4. STl per Game (mean=0.6, max=2)        counts: 10%
5. BLK per Game (mean=0.4, max=2.4)       counts: 5%
6. Field Goal  (combination of FGM & FG%)               counts: 20%
7.3 Point  (combination of 3PM & 3P%)              counts: 10%
8. Free Throw  (combination of FTM & FT%)                counts: 10%
9. Minutes played (the best abilitiy is availability)(mena=500, max=1000) counts: 5%
10. Games played (the best abilitiy is availability) (mean=24, max=32) counts: 5%

Idea: 
Manipulate all metrics so mean is 5. So all metrics have the same "value".
Then multiply by 2 for 20% by 0.5% for 5% etc.

```

```python
data2=data
data2.describe()
```

```python
fac=data2["PpG"].mean()/5
data2["PpG"]=data2["PpG"]/fac

fac1=data2["ApG"].mean()/5
data2["ApG"]=data2["ApG"]/fac1

fac2=data2["RpG"].mean()/5
data2["RpG"]=data2["RpG"]/fac2

fac3=data2["SpG"].mean()/5
data2["SpG"]=data2["SpG"]/fac3

fac4=data2["BpG"].mean()/5
data2["BpG"]=data2["BpG"]/fac4

data2["FG"]=data2["FG%"]*3 + data2["FGM"]
fac5=data2["FG"].mean()/5
data2["FG"]=data2["FG"]/fac5

data2["3P"]=data2["3P%"] + data2["3PM"]
fac6=data2["3P"].mean()/5
data2["3P"]=data2["3P"]/fac6

data2["FT"]=data2["FT%"] + data2["FTM"]
fac7=data2["FT"].mean()/5
data2["FT"]=data2["FT"]/fac7


fac8=data2["MIN"].mean()/5
data2["MIN"]=data2["MIN"]/fac8

fac9=data2["Games Played"].mean()/5
data2["Games Played"]=data2["Games Played"]/fac9

```

```python
data2["IER"]=data2["PpG"]*2 + data2["ApG"]*0.5 + data2["RpG"] + data2["SpG"] + data2["BpG"]*0.5 + data2["FG"]*2 + data2["3P"] + data2["FT"] + data2["MIN"]*0.5 + data2["Games Played"]* 0.5 /10
data2.sort_values('IER', ascending=False)
```
